```python
#Exercise 4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("StudentMentalhealth.csv")

# Convert Yes/No columns to numeric (0/1)
df['Depression_numeric'] = (df['Do you have Depression?'] == 'Yes').astype(int)
df['Anxiety_numeric'] = (df['Do you have Anxiety?'] == 'Yes').astype(int)
df['Panic_numeric'] = (df['Do you have Panic attack?'] == 'Yes').astype(int)

# Select numeric columns only (update column names to match your CSV)
subset = df[["Age", "What is your CGPA?", "Depression_numeric", "Anxiety_numeric", "Panic_numeric"]]

# Create a pairplot with hue
sns.pairplot(subset, hue="Depression_numeric", diag_kind="hist", palette="husl")

# Display the plot
plt.show()
```
