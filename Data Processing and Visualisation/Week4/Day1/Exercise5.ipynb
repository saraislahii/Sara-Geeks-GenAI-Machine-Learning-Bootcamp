```python
import matplotlib.pyplot as plt
# 1- Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
#-------------------------------------------------------------------------------
# Create x values manually (integers from -10 to 10)
x = list(range(-10, 11))

# Compute y = x^2
y = [i**2 for i in x]

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = xÂ²")

# Display the plot
plt.show()
```

```python
# 2- Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
#-------------------------------------------------------------------------------
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]  
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color=['grey', 'skyblue', 'lightgreen', 'pink'])
plt.xlabel("Products") 
plt.ylabel("Sales")
plt.title("Weekly Sales Data of Products")
plt.show()
```

```python
# 3- Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.
#-------------------------------------------------------------------------------
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]
colors = ['red', 'yellow', 'darkred', 'brown']
plt.figure(figsize=(8, 5))
plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%')
plt.title("Favorite Fruits Distribution")
plt.legend(fruits, title="Fruits", loc="best")
plt.show()
```
