```python
print("Exercise 3 : Visualizing Monthly Sales with a Bar Chart")
import matplotlib.pyplot as plt
months = ['January', 'February', 'March', 'April', 'May']
sales = [5000, 5500, 6200, 7000, 7500]
plt.bar(months, sales, color='blue')
plt.xlabel('Months')
plt.ylabel('Sales ($)')
plt.title('Monthly Sales Data')
plt.show()
```
