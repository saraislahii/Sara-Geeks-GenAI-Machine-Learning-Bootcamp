```python
import matplotlib.pyplot as plt
import pandas as pd

print("Exercise 4 : Data Visualisation")
# load dataset
data = pd.read_csv('Sales_data.csv')    # ensure this file exists in the same folder
df = pd.DataFrame(data)
print(df.head())



a= data['quantity'].sum()
print("Total quantity sold:", a)


# Calculate revenue by category
revenue_by_category = data.groupby('category')['revenue'].sum()

# Find category with highest revenue
highest_revenue_category = revenue_by_category.idxmax()
highest_revenue_amount = revenue_by_category.max()

print(f"\nCategory with highest revenue: {highest_revenue_category}")
print(f"Revenue generated: ${highest_revenue_amount:,.2f}")

print("\nRevenue by category:")
for category, revenue in revenue_by_category.items():
    print(f"{category}: ${revenue:,.2f}")
    
average= m= data['revenue'].mean()
print(f"Average revenue per sale: {average:.2f}")

# 4. Bar chart of total revenue per quarter 
# Ensure date column is datetime 
data['date'] = pd.to_datetime(data['date']) 
# Extract quarter
data['quarter'] = data['date'].dt.quarter 
# Sum revenue by quarter 
quarter_revenue = data.groupby('quarter')['revenue'].sum() 
# Plot bar chart 
plt.figure(figsize=(8, 5)) 
quarter_revenue.plot(kind='bar') 
plt.title("Total Revenue per Quarter")
plt.xlabel("Quarter (Q1â€“Q4)")
plt.ylabel("Revenue") 
plt.xticks([0,1,2,3], ["Q1", "Q2", "Q3", "Q4"], rotation=0)
plt.show()

```
